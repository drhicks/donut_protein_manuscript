```python
%matplotlib inline
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns

import matplotlib as mpl
mpl.rcParams.update({'font.size': 14})
```

```python
#get data
df = pd.read_csv("/home/drhicks1/donut_protein_manuscript/pocket_features/all.sc", sep="\s+")
df2 = pd.read_csv("/home/drhicks1/donut_protein_manuscript/pocket_features/ordered.sc", sep="\s+")
```

```python
from matplotlib.legend_handler import HandlerLine2D

for term in ["dimension_1", "dimension_2", "dimension_3", "pocket_volume"]:
    plt.figure()
    
    bp = plt.boxplot(df[term])
    plt.xlabel(term)
    
    y1 = df[term].values
    x1 = np.random.normal(1, 0.02, size=len(y1))
    
    y2 = df2[term].values
    x2 = np.random.normal(1, 0.02, size=len(y2))
    
    plt1, = plt.plot(x1, y1, 'b.', alpha=0.2, marker='o', label='scaffolds')
    plt2, = plt.plot(x2, y2, 'g.', alpha=0.95, marker='o', label='ordered')
    
    plt.legend(["scaffolds", "ordered"])    
    plt.legend(handler_map={plt1: HandlerLine2D(numpoints=1), plt2: HandlerLine2D(numpoints=1)})
    plt.tight_layout()
    plt.savefig(f"/home/drhicks1/donut_protein_manuscript/figs/{term}.png", dpi=400)
    plt.yscale("log")
    plt.savefig(f"/home/drhicks1/donut_protein_manuscript/figs/{term}_log.png", dpi=400)
    plt.show()
```

```python
from matplotlib.legend_handler import HandlerLine2D

for term in ["dimension_1", "dimension_2", "dimension_3", "pocket_volume"]:
    plt.figure()
    
    sns.distplot(df[term], kde=True)
    sns.distplot(df2[term], kde=True)
    plt.legend(["scaffolds", "ordered"])    
    plt.show()
```
